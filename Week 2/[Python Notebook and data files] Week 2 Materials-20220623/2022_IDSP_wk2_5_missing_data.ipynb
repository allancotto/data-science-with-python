We are going to use the ``` .copy()``` function to create a subset of the 'vechile_mpg' dataframe. Everything before the ``` .copy()``` command is creating a **view** of the data.  This is not an actual new bit of memory, rather we are telling the computer to look at a specific sub-section of the memory allocated the the eintire 'vichile_mpg' dataframe. From this point forward we want to modify and change the new dataframe that we are creating, not the orignal dataset. To do that we need to **copy** a selection of 'vichile_mpg' and place it in a new bit of memory. This highlights the object oriented nature of Python; we can have two objects looking at the same bit of memory. So if we change one, it is possiable to change the other. To avoid this we need to allocate more memory.

A good discussion of this can be found here:

https://realpython.com/pandas-settingwithcopywarning/

## 2.0 Finding the missing data

**Observe** that both the 'cylinders'  and 'displ' (i.e. engine displacement) series, only have a count of 44415 elements, not the 44787 rows that the full data set.  This simple tool has told us that for the data we have possesses missing values, but **where** and **Why**

To answer that lets introduce a new function ```.isna()```. This will search through the data and tell use where the missing values are. For the rest of the example I am going to focus on the 'cylinders' series. You will need to repeat for the 'displ' series.

#### 2.0.1 Example of the ``` .isna()``` function

````python
simple_mpg['cylinders'].isna()
````

We can see that this function produces a boolean series. But due to the length (44787 rows). we will have no clue as to where these missing values are. slicing out specific rows where the boolean value is 'True' should illuminate some things. 




#### 2.0.2 Example Use the ``` .loc ``` function to examine the rows missing data.


````python
simple_mpg.loc[simple_mpg['cylinders'].isna()]
````

### 2.1 Replacing missing values

Now can observe that the the missing values in the 'cylinders' column is due to these being electric cars.

At this point there are several options for how you can choose to deal with the missing data.  This decision will depend on the anlysis that you wish to persue, and the type of data that you have.  The possiable options are



1.   **Remove** all the rows with missing data using the ``.dropna()`` function.
2.    **Fill in**  missing values with specific value using ``.fillna()``
3.   **Interpolate** the missing values based on the rest of the data with ``.interpolate()``

All of these are valid options depending on the goals of your analysis.  Here we are just going to replace the NaN values with a fixed value of 0 since, electric vechiles have 0 cylinders.


#### 2.1.1 Example of filling in values

````python
simple_mpg['cylinders'].fillna(0)
````

Now re-examining the data 

````python
simple_mpg.loc[simple_mpg['cylinders'].isna() ]
````

What happened? the function ran with out error, but the orignal dataframe is unchanged. In this case, it should be pointed out that many of the native pandas functions do not overwirte or **mutate** the orignal object. Hence, when you apply the function you end up creating a new bit of memory. (In this case it is a new, but unnamed series.)

To address this we use the **keyword parameter** 'inplace' and set it equal to True (note capital T).

#### 2.1.2 Example of using inplace to chage orignal dataframe column

````python
simple_mpg['cylinders'].fillna(0,inplace=True)
````

````python
simple_mpg.loc[simple_mpg['cylinders'].isna()]
````

Again the filtering above now prodices an empty dataframe, which makes sense as we have filled in the cylinders coulmn.  But we have not affected the 'displ' column.  That is left for you to deal with in the next set of actions.  But first lets resave our data frame.
